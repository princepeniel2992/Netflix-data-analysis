# Queries Highlights
## Content Type Distribution: What is the distribution of content types (Movies vs. TV Shows) on Netflix?

```python
import matplotlib.pyplot as plt
```

```python
type_counts = netflix_df['type'].value_counts()

plt.bar(type_counts.index, type_counts.values, color=['orange', 'teal'])
plt.xlabel('Type')
plt.ylabel('Count')
plt.title('Count of Movies and TV shows')
plt.show()
```

## Top Directors: Who are the top 5 directors with the most content on Netflix?

```python
director_counts = netflix_df['director'].value_counts().head(5)

plt.bar(director_counts.index, director_counts.values)
plt.xlabel("Director")
plt.ylabel("Count")
plt.title("Count of Shows by top 5 Directors")
plt.xticks(rotation=45, ha='right')
plt.show()
```

## Cast Frequency: Which actors/actresses appear most frequently in Netflix titles?

```python
all_actors = []
for cast in netflix_df['cast'].dropna():
    all_actors.extend(cast.split(', '))


actor_counts = pd.Series(all_actors).value_counts().head(10)

#visulization part
plt.barh(actor_counts.index, actor_counts.values)
plt.xlabel("Actors/Actress")
plt.ylabel("Appearance Count")
plt.title("1o most frequent Actor/Actress appearance in Netflix titles")
plt.text(0,-0.1,"*Data as per 2021", ha='left', va='top', transform=plt.gca().transAxes)

plt.show()
```

## Yearly Content Addition: How has the number of titles added to Netflix changed over the years?

```python
#correct date time format
netflix_df['date_added'] = pd.to_datetime(netflix_df['date_added'], errors = 'coerce')
```

```python
#extract the year from 'date-added'
netflix_df['year_added'] = netflix_df['date_added'].dt.year
```

```python
#group bt 'year_added' and .size() counts all the releases in that particular year 
#including the null values if there are
yearly_content = netflix_df.groupby('year_added').size()
```

```python
# sort the values by year
yearly_content = yearly_content.sort_index()
```

```python
#plot the yearly content releases
plt.plot(yearly_content.index, yearly_content.values, marker='o')

plt.title('Yearly content addition over the years')
plt.xlabel('Year')
plt.ylabel('Releases')
plt.xticks(rotation = 45)

plt.show()
```

## Top Genres: What are the top 5 genres (from the ‘listed_in’ column) on Netflix?

```python
#step 1: split the 'listed_in' column and create a new dataframe where each genre is in separate row
genres_split = netflix_df['listed_in'].str.split(', ', expand=True).stack()

#step 2: count the occurence of each genre
top_genres = genres_split.value_counts()

#step 3: Get the top 5 genres
top_10_genres = top_genres.head(10)

print(top_10_genres)
```

```python
plt.bar(top_10_genres.index, top_10_genres.values, color="Maroon")
plt.title("Top 10 Genres based on the total releases")
plt.xlabel("Genres")
plt.ylabel('Total releases')
plt.xticks(rotation = 90)

plt.show()
```
